### 1. Effect of no. of neurons on Linear Autoencoder 

The number of neurons did not significantly change for the Linear Autoencoder. By doubling the neurons, the autoencoder's RMSE was approximately the same.

### 2. Effect of no. of neurons on Non-Linear Autoencoder

It is clear that in the instance of the Non-Linear Autoencoder, adding more neurons led to a higher-quality decoding and reconstruction of the data. For a dataset size of 999000, the latent space (encoding dimension) was 2 and the input variables were 4. Autoencoder performed poorly in compression and reconstruction when compared to PCA since there were very input variables compared to the size of dataset. In situations involving higher input variables, such as picture compression, autoencoders outshine PCA.

### 3. Effect of Activation function on Autoencoders

It was discovered that Swish and Relu outperformed other activation functions by maintaining same hyperparameters. Linear was the worst activation function. Higher order optimization considerably decreased the root mean square error (RMSE) for the Tanh activation function. Forst All orders of Optimization, adam optimizer was used.

### 4. Latent Space same as Input Layer

Since 4 is a very little input compared to the size of the dataset, the input layer and latent space structure were left same.

Although a lossless copy of the input data should be produced ideally, there is a small amount of reconstruction error with non-linear autoencoders. There is a significant data loss during copying for Linear Autoencoder.
